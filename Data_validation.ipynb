```python
import os

dataset_path = "plantvillage dataset/color"

for disease in os.listdir(dataset_path):
    disease_path = os.path.join(dataset_path, disease)
    
    if os.path.isdir(disease_path):
        num_images = len(os.listdir(disease_path))
        print(disease, ":", num_images)
```

```python
from PIL import Image
import os

dataset_path = "plantvillage dataset/color"

for root, dirs, files in os.walk(dataset_path):
    
    for file in files:
        
        path = os.path.join(root, file)
        
        try:
            img = Image.open(path)
            img.verify()
        except:
            print("Removing corrupted image:", path)
            os.remove(path)
```

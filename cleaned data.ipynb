```python
import os
import cv2
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
```

```python
dataset_path = "/kaggle/input/datasets/abdallahalidev/plantvillage-dataset"
output_path = "/kaggle/working/clean_dataset"

os.makedirs(output_path, exist_ok=True)
```

```python
image_paths = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        image_paths.append(os.path.join(root, file))

print("Total images:", len(image_paths))
```

```python
def process_image(path):

    try:
        img = cv2.imread(path)

        if img is None:
            return

        img = cv2.resize(img,(224,224))

        class_name = path.split("/")[-2]

        save_dir = os.path.join(output_path,class_name)
        os.makedirs(save_dir,exist_ok=True)

        filename = os.path.basename(path)

        cv2.imwrite(os.path.join(save_dir,filename),img)

    except:
        pass
```

```python
with ThreadPoolExecutor(max_workers=16) as executor:
    list(tqdm(executor.map(process_image,image_paths), total=len(image_paths)))
```

```python
count = 0

for root,dirs,files in os.walk(output_path):
    count += len(files)

print("Clean images:",count)
```

```python
import shutil

shutil.make_archive(
    "/kaggle/working/clean_plantvillage",
    'zip',
    output_path
)
```

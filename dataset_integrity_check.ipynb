```python
# Dataset Validation & Label Verification
# Role: Dataset Validation Engineer

import os
from PIL import Image
import matplotlib.pyplot as plt

print("Dataset integrity validation started...")
```

```python
dataset_path = "my_crop_data/plantvillage/plantvillage dataset/color"
```

```python
# Count number of images in each class folder

class_counts = {}

for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder):
        images = os.listdir(class_folder)
        class_counts[class_name] = len(images)

# Print results
for cls, count in class_counts.items():
    print(f"{cls} : {count} images")
```

```python
# Detect corrupted or unreadable images

corrupted_images = []

for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder):
        for img_file in os.listdir(class_folder):
            img_path = os.path.join(class_folder, img_file)

            try:
                img = Image.open(img_path)
                img.verify()  # check if image is readable
            except:
                corrupted_images.append(img_path)

print("Total corrupted images found:", len(corrupted_images))
```

```python
# Faster corrupted image detection
import os
from PIL import Image
dataset_path = "my_crop_data/plantvillage/plantvillage dataset/color"
corrupted_images = []
checked = 0

for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder):
        for img_file in os.listdir(class_folder):
            img_path = os.path.join(class_folder, img_file)
            checked += 1

            try:
                with Image.open(img_path) as img:
                    img.verify()
            except:
                corrupted_images.append(img_path)

            # progress update every 1000 images
            if checked % 1000 == 0:
                print(f"Checked {checked} images...")

print("Total images checked:", checked)
print("Total corrupted images found:", len(corrupted_images))
```

```python
print("DATASET VALIDATION SUMMARY")
print("--------------------------")
print("Dataset path:", dataset_path)
print("Total images checked:", checked)
print("Total corrupted images found:", len(corrupted_images))

if len(corrupted_images) == 0:
    print("Dataset integrity status: PASSED (no corrupted images)")
else:
    print("Dataset integrity status: ISSUES FOUND")
```
